```python
import pandas as pd
import numpy as np
from xgboost import XGBClassifier
from sklearn.model_selection import cross_val_score
from imblearn.over_sampling import SMOTE
import warnings
from sklearn.model_selection import KFold
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
warnings.filterwarnings("ignore")
```

```python
def execute_eval_HT(path_fake, path_real, seed):
    df_real = pd.read_csv(path_real)
    df_real['label'] = 0
    df_fake = pd.read_csv(path_fake)
    df_fake['label'] = 1
    df_fake_real = pd.concat([df_fake, df_real])
    train_X = df_fake_real.loc[:, 'bias':'state']
    train_y = df_fake_real['label']
    sm = SMOTE(random_state=42)
    train_X, train_y = sm.fit_resample(train_X, train_y)
    strat_k_fold = KFold(n_splits=5, shuffle=True, random_state=seed)
    model = XGBClassifier(verbosity = 0)
    results_roc = cross_val_score(model, train_X, train_y, cv=strat_k_fold, scoring='roc_auc')
    results_precision = cross_val_score(model, train_X, train_y, cv=strat_k_fold, scoring='precision')
    results_recall = cross_val_score(model, train_X, train_y, cv=strat_k_fold, scoring='recall')
    results_f1 = cross_val_score(model, train_X, train_y, cv=strat_k_fold, scoring='f1')
    return results_roc, results_precision, results_recall, results_f1
```

```python
def random_sampling_eval(path_fake, path_real, seed):
    df_real = pd.read_csv(path_real)
    df_real['label'] = 0
    df_fake = pd.read_csv(path_fake)
    df_fake['label'] = 1
    df_fake_real = pd.concat([df_fake, df_real])
    X = df_fake_real.loc[:, 'bias':'state']
    y = df_fake_real['label']
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=seed)
    sm = SMOTE(random_state=42)
    X_train, y_train = sm.fit_resample(X_train, y_train)
    model = XGBClassifier(verbosity = 0)
    model.fit(X_train, y_train)
    pred_results = model.predict(X_test)
    
    f1_val = f1_score(y_test, pred_results, average='binary')
    precision = precision_score(y_test, pred_results, average='binary')
    recall = recall_score(y_test, pred_results, average='binary')
    roc_auc = roc_auc_score(y_test, model.predict_proba(X_test)[:, 1])
    return roc_auc, f1_val, precision, recall
    
```

```python
import shap
shap.initjs()
```

```python
def SHAP_analysis(path_real, path_fake):
    df_real = pd.read_csv(path_real)
    df_real['label'] = 0
    df_fake = pd.read_csv(path_fake)
    df_fake['label'] = 1
    df_fake_real = pd.concat([df_fake, df_real])
    train_X = df_fake_real.loc[:, 'bias':'state']
    train_y = df_fake_real['label']
    model = XGBClassifier(n_jobs=4)
    model.fit(train_X, train_y)
    X_train_arr = train_X
    shap_values = shap.TreeExplainer(model).shap_values(X_train_arr)
    shap.summary_plot(shap_values, X_train_arr, feature_names=['bias', 'conspiracy', 'hate', 'satire', 'junk', 'state'], plot_type='bar')
    shap.summary_plot(shap_values, X_train_arr, feature_names=['bias', 'conspiracy', 'hate', 'satire', 'junk', 'state'], plot_type='dot')
```

## SHAP analysis of model trained with BS dataset

```python
SHAP_analysis('BS_top_real_reduce.csv', 'BS_top_fake_reduce.csv')
```

## SHAP analysis of model trained with COVID data

```python
SHAP_analysis('COVID_top_real_reduce.csv', 'COVID_top_fake_reduce.csv')
```

## SHAP analysis of model trained with CELEBRITY data

```python
SHAP_analysis('CELEBRITY_top_real_reduce.csv', 'CELEBRITY_top_fake_reduce.csv')
```
